```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Importing the Big Basket Cart Dataset

```python
dataset=pd.read_csv('BigBasketCart.csv', header=None)
```

Adding Rows and Columns from the Dataset to the transaction variable   

```python
transactions=[]
for i in range(0,7219):
  transactions.append([str(dataset.values[i,j]) for j in range(0,20)])
```

Importing Apriori and training Apriori using defined sample params

```python
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

Verifying the result

```python
results = list(rules)
```

```python
results
```

```python
print(results)
```

Converting the Raw result to proper format

```python
def inspect(results):
    product1         = [tuple(result[2][0][0])[0] for result in results]
    product2         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(product1, product2, supports, confidences, lifts))
DataFrame_intelligence = pd.DataFrame(inspect(results), columns = ['product1', 'product1', 'Support', 'Confidence', 'Lift'])
```

Displaying the Final Cart Prediction

```python
DataFrame_intelligence
```

Prioritizing items that have higher lift

```python
DataFrame_intelligence.nlargest(n = 10, columns = 'Lift')
```
